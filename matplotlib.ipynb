```python
import matplotlib.pyplot as plt
```

```python
plt.show()
```

```python
import numpy as np

x=np.linspace(0,5,11)
y =x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'y-')
plt.show()

plt.xlabel('X Label')
plt.ylabel('y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig=plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
```

```python
fig = plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.17,0.4,0.5])

axes1.plot(x,y)
axes.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig=plt.figure(figsize=(8,2))
ax=fig.add_axes([0,0,1,1,])
ax.plot(x,y)
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='x Squared')
ax.plot(x,x**3,label='x Cubed')

ax.legend(loc=10)
```
